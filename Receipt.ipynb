```python
import datetime
from prettytable import PrettyTable
from tabulate import tabulate
dt=datetime.datetime.now()
data=[]
Items={}
Rate_Qty={}
Rate_Value=[]
Total_Amt=0
Total_Qty=0
Total_Rate=0  
line=" _ "*20
File_End=open("End.txt","r")
#Company Info
Market_Name,Location="BHAT-BHATENI","ITAHARI-SUNSARI"
Discount_Percentage,Payment_Mode=float(input("Discount % ")),"Cash"
#Customer Info
usr=(
input("Name: "),int(input("Pan No: ")),
float(input("Customer Money: ")),int(input("Vat Number: ")))
data.append(usr)
Customer_Name,p_no,Customer_Money,Vat_No=usr[0],usr[1],usr[2],usr[3]
#Dictonary of rate and quantity with item name
Total_Items=int(input("Total Items: "))
for i in range(Total_Items):
    Items_Name=input("Item Name: ")
    Item_Qty=int(input("Item Quantity: "))
    Item_Rate=int(input("Item Rate: "))
    Rate_Qty[Item_Qty]=Item_Rate
    Items[Items_Name]=Rate_Qty
#Value Extraction as list
Items_Keys=list(Items.keys())
Items_Values=list(Items.values())
for i in range(len(Items_Keys)):
    Values_Items=Items_Values[i]
    Quantity_Values=list(Values_Items.keys())
    Rate_Value=list(Values_Items.values())
Quantity=list(map(int,Quantity_Values))
for i in range(len(Values_Items)):
    Total_Amt+=((Quantity_Values[i])*(Rate_Value[i]))
    Total_Qty+=(Quantity_Values[i])
    Total_Rate+=(Rate_Value[i])
Discount_Amt=((Discount_Percentage/100)*(Total_Amt))
Net_Amount=Total_Amt-Discount_Amt
Change=Customer_Money-Net_Amount
print(
"\t"*4,Market_Name,"SUPER MARKET","(",Location,")","\n",
"\t   "*5,Location,"\n",
"\t"*5,"   VAT No:",Vat_No,"\n",
"\t"*5,"ABBREVIATED TAX INVOICE","\n")
Info_Data=[
    ["\tBill               :","SI3252-ITH-062/53"],
    ["\tTransaction Date   :",dt],
    ["\tBill To            :",Customer_Name],
    ["\tPan No             :",p_no],
    ["\tPayment Mode       :",Payment_Mode]
]
Data_Cash=[
    ['\tGross Amount       :',Total_Amt],
    ['\tDiscount           :',Discount_Amt],
    ["\tTender             :",Customer_Money],
    ["\tChange             :",Change],
    ['\tNet Amount         :',Net_Amount]
]
Data_Item=[
    ["\tParticulars        :"," , ".join(map(str,Items_Keys))],
    ["\tQuantity           :"," , ".join(map(str,Quantity_Values))],
    ["\tRate               :"," , ".join(map(str,Rate_Value))],
]
print("Transaction Details")
print(tabulate(Info_Data))
print("Item Details")
print(tabulate(Data_Item))
print("Finance Details")
print(tabulate(Data_Cash))
print(File_End.read())
File_End.close()
```
